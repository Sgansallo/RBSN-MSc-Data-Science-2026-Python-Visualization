# Introduction to Python for Data Science

### Making a mental shift from Microsoft Excel to Python for Data Science

In Excel, you have 'cells'. A cell is at the intersection of rows and columns. Cells are named according to the row and column where it intersected. For instance, a cell at the intersection of Column A and Row 1 is named 'A1'. And so, we can write anything (text, number, alphanumeric, etc) in A1. Technically, the name A1 is used to refer to the item contained in the cell. It's same in Python. We can use names like A1, etc to refer to any item we want, if we want to maintain the mental model of working in Excel.

````python
b6 = "a"
c6 = 5
````

However, there is a better way to work in Excel. As seen in the image below, we can use Excel's name manager to name whatever cell (or group of cells) we work with. We can use the name 'height' to represent/refer to the items in `Cell A1:A3`. This is actually what we do when we use variables in Python: we use a name to refer to a group of items in our computer's memory (RAM). So in Python, we'll write:
```python
height = [12, 13, 15]
```

image.png

But the journey from Excel to Python is not really straightforward, as there are some surprising behaviour/quirks in Python, different than what we know in Microsoft Excel. For instance, in Excel, if we have `a` in cell `B6` and the number `5` in Cell `C6`, multiplying `B6 * C6` in Excel will give us a `#VALUE` error, but as you can see below, it works perfectly in Python.

````python
print(b6 * c6)
````

We can add data into Python using a data structure called a `list`  like below. We'll talk more about this soon

````python
height = [1, 2, 3]
weight = [1, 2, 3]
````

If we used Excel's name manager to name the ranges we want to work with, we'll see that Excel looks quite similar to Python (a little bit). For instance, in the image below, we can use Excel's SUM function to get the total of the data in the range named `height`

````python
def calculate_age(this_year, birth_year):
  """
  This function calculates the age of a person

  Parameters:
  this_year (int): The current year
  birth_year (int): The year of birth

  Returns:
  age (int): The age of the person

  """
  age = this_year - birth_year
  return age
````
